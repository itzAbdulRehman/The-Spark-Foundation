```python
plt.figure(figsize=(11,6))
team = match[['team1']]
team = match[['team2']]
ax = sns.countplot(pd.concat([team,team],ignore_index=True)['team2'])
for p in ax.patches:
    ax.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01),ha="center")
plt.xticks(rotation=90)
plt.show()
```
